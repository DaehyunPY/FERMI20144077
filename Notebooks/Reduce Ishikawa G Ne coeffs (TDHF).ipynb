```python
import pandas as pd
import numpy as np
```

```python
sheets = pd.read_excel("Data/Simulated Ne beta parameters, TDHF, 2018-09-17.xlsx", sheet_name=None, index_col=0)
```

```python
def cmpx(amp, shift):
    return amp * np.exp(1j * shift)


def split_p(data: pd.DataFrame, k: pd.DataFrame):
    p = cmpx(data["p_amp"], data["p_shift"])
    pdp1 = p.loc[1]
    pdp0 = k.loc[0, "pdp_amp"] * pdp1 / k.loc[1, "pdp_amp"]
    psp0 = p.loc[0] - pdp0
    ret = pd.DataFrame({
        "pdp": {-1: pdp1, 0: pdp0, 1: pdp1},
        "psp": {-1: np.nan, 0: psp0, 1: np.nan},
    }).rename_axis('m')
    return pd.concat([
        data.drop(columns=['p_amp', 'p_shift']),
        ret.apply(np.abs).add_suffix("_amp"),
        ret.apply(np.angle).add_suffix("_shift"),
    ], axis=1)
```

```python
key = "good1"
splited = split_p(sheets[key], sheets["k"])
(splited/sheets["k"]).combine_first(splited)
```
