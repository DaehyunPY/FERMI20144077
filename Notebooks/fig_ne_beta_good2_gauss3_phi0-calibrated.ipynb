```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.read_excel('Data/beta_good2_gauss3.xlsx', sheet_name=None)

for k, d in df.items():
    print(f'sheet: {k}')
    display(d)
```

```python
def err_operadd(a, b):
    return (a**2 + b**2) ** 0.5


phi0 = 5.101
df['data']['he_beta1m3_opt'] = df['data']['he_beta1_opt'] - 2/3 * df['data']['he_beta3_opt']
df['data']['he_beta1m3_err'] = err_operadd(df['data']['he_beta1_err'], 2/3 * df['data']['he_beta3_err'])
df['data']['ne_beta1m3_opt'] = df['data']['ne_beta1_opt'] - 3/2 * df['data']['ne_beta3_opt']
df['data']['ne_beta1m3_err'] = err_operadd(df['data']['ne_beta1_err'], 3/2 * df['data']['ne_beta3_err'])
df['data']['phase_adj'] = df['data']['phase']*2*np.pi - phi0
df['data'] = df['data'].set_index('phase_adj').sort_index()
df['data']
```

```python
th = np.linspace(-5, 6, 111)
props = iter(plt.rcParams['axes.prop_cycle'])
d = df['data']
f = dict(df['fitted'].iloc[0])

plt.style.use('Science.mplstyle')
plt.figure(figsize=(2.25, 2.25))
plt.xlabel(r'$\omega$-$2\omega$ relative phase (rad.)')
plt.ylabel(r'$\beta$ parameters')

for key, label in [['ne_beta1', r'$\beta_1$'],
                   ['ne_beta2', r'$\beta_2$'],
                   ['ne_beta1m3', r'$\beta_1 - 3/2 \beta_3$'],
                   ['ne_beta4', r'$\beta_4$']]:
    prop = next(props)
    plt.errorbar(d.index, d[key], yerr=d[f'{key}_err'],
                 fmt='o', label=label, **prop)
    if key in {'ne_beta1', 'ne_beta3', 'ne_beta1m3'}:
        plt.plot(th, (f[f'{key}_amp_value']
                      * np.cos(th-f[f'{key}_shift_value']+phi0)
                      + f[f'{key}_offset_value']),
                 **prop)
    elif key in {'ne_beta2', 'ne_beta4'}:
        plt.plot(th, 0*th + f[f'{key}_c_value'],
                 **prop)

xlim = plt.xlim()
plt.locator_params(axis='x', steps=[1, 2, 10])
plt.locator_params(axis='y', steps=[1, 2, 10])
plt.grid(True)

plt.twiny()
plt.xlim(*xlim)
prop = {'color': 'k', 'linestyle': '--'}
vlines = [
    {'at': (f[f'ne_beta1_shift_value']-phi0)%(2*np.pi),
     'rough': -1.3,
     'label': r'$\eta_f-\eta_d$'},
    {'at': (f[f'ne_beta1m3_shift_value']-phi0)%(2*np.pi),
     'rough': 2.2,
     'label': r'$\eta_f-\eta_s$'},
]
for v in vlines:
    plt.axvline(v['at'], **prop)
plt.xticks([v['rough'] for v in vlines], [v['label'] for v in vlines], ha='left')
plt.tick_params(axis='x', which='both', length=0)

plt.tight_layout(pad=0)
plt.savefig('Figures/fig_ne_beta_good2_gauss3.pdf')
plt.show()
```
