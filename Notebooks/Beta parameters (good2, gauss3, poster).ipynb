```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.read_excel('Data/beta_good2_gauss3.xlsx', sheet_name=None)

for k, d in df.items():
    print(f'sheet: {k}')
    display(d)
```

```python
def err_operadd(a, b):
    return (a**2 + b**2) ** 0.5


df['data']['he_beta1m3_opt'] = df['data']['he_beta1_opt'] - 2/3 * df['data']['he_beta3_opt']
df['data']['he_beta1m3_err'] = err_operadd(df['data']['he_beta1_err'], 2/3 * df['data']['he_beta3_err'])
df['data']['ne_beta1m3_opt'] = df['data']['ne_beta1_opt'] - 3/2 * df['data']['ne_beta3_opt']
df['data']['ne_beta1m3_err'] = err_operadd(df['data']['ne_beta1_err'], 3/2 * df['data']['ne_beta3_err'])
```

```python
th = np.linspace(0, 11, 111)
phi0 = 5.101
d = df['data']
f = dict(df['fitted'].iloc[0])

plt.style.use('My Poster.mplstyle')
plt.figure(figsize=(7.5, 12))
plt.subplot(211)
props = iter(plt.rcParams['axes.prop_cycle'])
plt.xlabel(r'$\omega$-$2\omega$ relative phase + $\phi_0$ (rad.)')
plt.ylabel(r'He $\beta$ parameters')

for key, label in [['he_beta1m3', r'$\beta_1 - 2/3 \beta_3$'],
                   ['he_beta2', r'$\beta_2$'],
                   ['he_beta3', r'$\beta_3$'],
                   ['he_beta4', r'$\beta_4$']]:
    prop = next(props)
    plt.errorbar(2*np.pi*d['phase'], d[key], yerr=d[f'{key}_err'],
                 fmt='o', label=label, **prop)
    if key in {'he_beta1m3', 'he_beta3'}:
        plt.plot(th, (f[f'{key}_amp_value']
                      * np.cos(th-f[f'{key}_shift_value'])
                      + f[f'{key}_offset_value']),
                 **prop)
    elif key in {'he_beta2', 'he_beta4'}:
        plt.plot(th, 0*th + f[f'{key}_c_value'],
                 **prop)

plt.xlim(0, 11)
plt.locator_params(axis='y', steps=[1, 5, 10])
plt.grid(True)
prop = {'color': 'k', 'linestyle': '--'}
vlines = [
    {'at': f[f'he_beta1m3_shift_value']%(2*np.pi) + 2*np.pi,
     'rough': 1,
     'label': r'$\eta_s-\eta_p+\phi_0$'},
    {'at': f[f'he_beta3_shift_value']%(2*np.pi) + np.pi,
     'rough': 3,
     'label': r'$\eta_d-\eta_p+\phi_0$'},
]

for v in vlines:
    plt.axvline(v['at'], **prop)
plt.gca().label_outer()

plt.subplot(212)
props = iter(plt.rcParams['axes.prop_cycle'])
plt.xlabel(r'$\omega$-$2\omega$ relative phase + $\phi_0$ (rad.)')
plt.ylabel(r'Ne $\beta$ parameters')

for key, label in [['ne_beta1', r'$\beta_1$'],
                   ['ne_beta2', r'$\beta_2$'],
                   ['ne_beta1m3', r'$\beta_1 - 3/2 \beta_3$'],
                   ['ne_beta4', r'$\beta_4$']]:
    prop = next(props)
    plt.errorbar(2*np.pi*d['phase'], d[key], yerr=d[f'{key}_err'],
                 fmt='o', label=label, **prop)
    if key in {'ne_beta1', 'ne_beta1m3'}:
        plt.plot(th, (f[f'{key}_amp_value']
                      * np.cos(th-f[f'{key}_shift_value'])
                      + f[f'{key}_offset_value']),
                 **prop)
    elif key in {'ne_beta2', 'ne_beta4'}:
        plt.plot(th, 0*th + f[f'{key}_c_value'],
                 **prop)

plt.xlim(0, 11)
plt.locator_params(axis='y', steps=[1, 2, 5, 10])
plt.grid(True)
prop = {'color': 'k', 'linestyle': '--'}
vlines = [
    {'at': f[f'ne_beta1_shift_value']%(2*np.pi) + 2*np.pi,
     'label': r'$\eta_s-\eta_p+\phi_0$'},
    {'at': f[f'ne_beta1m3_shift_value']%(2*np.pi) + 2*np.pi,
     'label': r'$\eta_d-\eta_p+\phi_0$'},
]

for v in vlines:
    plt.axvline(v['at'], **prop)

plt.tight_layout()
plt.savefig('Figures/Beta parameters (good2, gauss3, poster).pdf')
plt.show()
```
