```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lmfit.models import ExpressionModel
```

```python
sheets = pd.read_excel('Data/Simulated Ne beta parameters, TDCASSCF, 2018-10-04.xlsx', sheet_name=None)
sheets['w2w'].set_index(['target dataset', 'photon (eV)', 'phi (deg)'], inplace=True)
sheets['wonly'].set_index('photon (eV)', inplace=True)
```

```python
print("w2w experiment:")
sheets['w2w']
```

```python
print("w2w experiment (mean):")
sheets['w2w'].groupby('target dataset')[['beta2', 'beta4']].mean()
```

```python
print("wonly experiment:")
sheets['wonly']
```

```python
dataset = 'good1'
keys = ['beta1', 'beta3', 'beta1m3']

model = ExpressionModel('amp * cos(freq * x - shift) + offset')
model.set_param_hint('amp', value=1, min=0)
model.set_param_hint('freq', value=1, vary=False)
model.set_param_hint('shift', value=np.pi/2, min=-np.pi, max=3*np.pi)
model.set_param_hint('offset', value=0)

df = sheets['w2w'].loc[dataset]
df.index = df.index.droplevel('photon (eV)')
plt.figure()
for k, prop in zip(keys, plt.rcParams['axes.prop_cycle']):
    fitted = model.fit(df[k], x=df.index/180*np.pi)
    print(k)
    print(fitted.fit_report())
    th = np.linspace(0, 2*np.pi, 361)
    plt.plot(df.index/180*np.pi, df[k], '.--', **prop)
    plt.plot(th, fitted.eval(x=th), **prop)
plt.show()
```
